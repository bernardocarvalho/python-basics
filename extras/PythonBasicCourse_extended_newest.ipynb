# The SCIPY module - basic scientific methods

```python
# SciPy
from scipy.optimize import curve_fit
#DEfine function f(x) which has extra arguments a,b,c
def f(x,a,b,c):
    return a * np.exp(-b * x ) +c
n=60
x = np.linspace(0,5, n)
yclean = f(x, 5, 2, 0.5)
ynoisy = yclean + 0.5 * np.random.randn(n) #add normal gaussian noise mean=0, std_dev=1
```

### Basic interpolation

```python
from scipy.interpolate import UnivariateSpline
# Create the spline interpolant using the noisy data
spline_interp = UnivariateSpline(x, ynoisy)
# Plot the data and the fitted function
plt.scatter(x, ynoisy, label='Noisy Data', color='blue')
plt.plot(x, yclean, label='True Function', color='green', linestyle='dotted', linewidth=2)
plt.plot(x, spline_interp(x), label='Spline Interpolation', color='orange', linewidth=2)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve Fitting with Spline Interpolation')
plt.grid(True)
plt.show()
```

### Curve fitting

```python
popt, pcov = curve_fit (f, x, ynoisy) #provide the "model" function and X and Y data series
perr = np.sqrt(np.diag(pcov)) #calculate the "error" estimate in the fit parameters
y_fit = f(x, *popt) #calculate the array of fit signal values using the "best fit parameters"
msd = np.sum((ynoisy-y_fit) ** 2)/n #calculate the normalised sum of squared fit error
```

```python
#The true parameters are: 5, 2, 0.5
print(['{:.4f} +/- {:.4f}'.format(popt[i],perr[i]) for i in range(3)]) #use only 4 decimals
print('{:.4f}'.format(msd))
```

```python
plt.plot(x,ynoisy,'*',label='noisy')
plt.plot(x,y_fit,label='fit')
plt.plot(x,yclean,label='clean')
plt.legend(loc='best')
plt.show()
```

### Piecewise curve fitting

```python
# Define the two functions with two coefficients each
def func1(x, a, b):
    return a * x + b

def func2(x, c, d):
    return c * x + d

# Define the piecewise model with two functions and a single transition point
def piecewise_function(x, a, b, c, d, transition):
    y=func1(x,a,b)
    y[x > transition]=func2(x[x > transition],c,d)
    #print(transition)
    return y
    #return np.piecewise(x, [x <= transition, x > transition], [lambda x: func1(x, a, b), lambda x: func2(x, c, d)])

# Define the piecewise model with two functions and a single transition point
def piecewise_C2_function(x, a, b, c, d, transition):
    y=func1(x,a,b)*(1-sigmoid(x-transition,k=50)) + func2(x,c,d)*sigmoid(x-transition,k=50)
    return y

#######################################################    
# Generate example data with a single transition point
x_data = np.linspace(0, 10, 100)
transition_point = 3.5
#y_data = 2 * x_data + 1  # Function1 on the first segment
#y_data[x_data > transition_point] = -3 * x_data[x_data > transition_point] + 30  # Function2 on the second segment
y_data = piecewise_function(x_data, 2, 1, -3, 30, transition_point)

# Perform the curve fit
initial_guess = [1.0, 1.0, 1.0, 1.0, 4.5]  # Initial guess for the coefficients and transition point

# Define the bounds for the coefficients and transition point during the fit
lower_bounds = [-np.inf, -np.inf, -np.inf, -np.inf, 0.0]
upper_bounds = [np.inf, np.inf, np.inf, np.inf, 10.0]
popt, cov = curve_fit(piecewise_C2_function, x_data, y_data, p0=initial_guess, 
                    bounds=(lower_bounds, upper_bounds))

# Extract the coefficients and transition point
a, b, c, d, transition_point = popt

# Print the results
print("Coefficients for Function 1: a =", a, "b =", b)
print("Coefficients for Function 2: c =", c, "d =", d)
print("Transition point: x =", transition_point)

# Fit metrics
y_fit = piecewise_function(x_data, *popt)
residuals = y_data - y_fit
degrees_of_freedom = len(x_data) - len(popt)
chi_squared = np.sum(residuals**2)
reduced_chi_squared = chi_squared / degrees_of_freedom
print("Reduced Chi-Squared:", reduced_chi_squared)

# Plot the data and the fitted piecewise function
plt.scatter(x_data, y_data, label='Data', color='blue')
plt.plot(x_data, piecewise_function(x_data, a, b, c, d, transition_point), label='Piecewise Fit', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Piecewise Regression with Arbitrary Functions')
plt.grid(True)
plt.show()
```

### Cross correlation between signals

```python
import math
from scipy import signal
pi=math.pi

##########################
# Construction of both
#    Signals
##########################
n=1000
sfreq=1e3 #sampling frequency
dt=1/sfreq  #sampling time
t=np.arange(0,n)*dt
f1=30.0  #in Hz
y1 = 2.0*np.sin(2.0*pi*f1*t) + 0.2*np.random.randn(n)
y2 = 2.0*np.sin(2.0*pi*f1*t+pi/4.0) + 0.2*np.random.randn(n)
time_shift=(pi/4.0)/(2.0*pi*f1)

##########################
# Ccorrelation between 
# both signals
##########################
corr = signal.correlate(y1, y2,mode='same')/np.linalg.norm(y1)/np.linalg.norm(y2)
#Array of lags
lags = signal.correlation_lags(len(y1), len(y2),mode='same')*dt
max_corr=np.argmax(np.abs(corr)) #highest in magnitude is what we want...

##########################
# Plotting the result
##########################
fig, ax = plt.subplots(2)
ax[0].plot(t,y1,label='y1')
ax[0].plot(t,y2,label='y2')
ax[0].set_xlim([0,0.1])
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Amplitude [V]')
ax[0].legend(loc='best')
ax[0].set_title('correlated time series; time_shift='+str('%.5f'%time_shift))
ax[1].plot(lags,corr,label='corr')
ax[1].plot([lags[max_corr],lags[max_corr]],[-1,1],label='max_corr='+str('%.3f'%lags[max_corr]),color='red',linestyle='--')
ax[1].set_xlim([-0.05,0.05])
ax[1].set_xlabel('Time lag [s]')
ax[1].set_ylabel('Cross Corr. []')
ax[1].legend(loc='best')
fig.subplots_adjust(hspace=0.4)
plt.show()
```

```python
print(np.argmax(corr))
print(corr[np.argmax(corr)])
print(lags[np.argmax(corr)])
```

### Cross Spectral Density between signals

```python
# fft of y1 and the phase between the y1 and y2 at the signal frequency using the cross spectrum

fft_y1 = np.fft.fft(y1)
freqs = np.fft.fftfreq(n, dt)
print(freqs[int(n/2)-1])
signal_freq_index = np.argmax(np.abs(fft_y1[0:int(n/2)-1]))
signal_freq = freqs[signal_freq_index]

freqs_csd, csd = signal.csd(y1, y2, fs=sfreq, nperseg=n)
# Find the index of the signal frequency in the CSD frequency array
signal_freq_index_csd = np.argmin(np.abs(freqs_csd - signal_freq))
phase = np.angle(csd[signal_freq_index_csd], deg=True)

plt.figure()
plt.plot(freqs[0:int(n/2)-1], np.abs(fft_y1)[0:int(n/2)-1])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('FFT of y1')
plt.xlim([20, 40]) # Limit the x-axis to show the relevant frequencies
plt.show()

print(f"Signal frequency: {signal_freq:.2f} Hz")
print(f"Phase between y1 and y2 at signal frequency: {phase:.2f} degrees")
```
